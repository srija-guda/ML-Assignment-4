```python
def remove_outliers(df, column_name):
    mean = df[column_name].mean()
    std_dev = df[column_name].std()

    lower_bound = mean - 3 * std_dev
    upper_bound = mean + 3 * std_dev

    df_cleaned = df[(df[column_name] >= lower_bound) & (df[column_name] <= upper_bound)]

    return df_cleaned
```

```python
for col in df.columns:
    if (col == 'Outcome' or col == 'Insulin'):
      continue
    df = remove_outliers(df, col)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('Outcome', axis=1), df['Outcome'], test_size=0.2, random_state=0)
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
class CustomPerceptron:
    def __init__(self, learning_rate=0.01, n_iters=1000):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.weights = None
        self.bias = 0

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)

        y_ = np.where(y <= 0, -1, 1)

        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = np.sign(linear_output)

                if y_predicted != y_[idx]:
                    self.weights += self.learning_rate * y_[idx] * x_i
                    self.bias += self.learning_rate * y_[idx]

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = np.sign(linear_output)
        return np.where(y_predicted == -1, 0, 1)
```

```python
perceptron = CustomPerceptron(learning_rate=0.01, n_iters=1000)
perceptron.fit(X_train, y_train)
predictions = perceptron.predict(X_test)
accuracy3 = accuracy_score(y_test, predictions) *100
precision3 = precision_score(y_test, predictions) *100
recall3 = recall_score(y_test, predictions) *100
f1_score3 = f1_score(y_test, predictions) *100


print(f"Accuracy: {accuracy3:.3f}%")
print(f"Precision: {precision3:.3f}%")
print(f"Recall: {recall3:.3f}%")
print(f"F1 Score: {f1_score3:.3f}%")
```

Naive bayes model has a slightly better accuracy than perceptron, indicating it is slightly more reliable than perceptron in identifying correct class. Naive bayes also has a higher precision than perceptron, indicating that it is better in identifying positive classes while generating fewer false positives. This is useful when we want fewer false positives. The recall for the perceptron is slightly better, indicating that it is slightly better at identifying positive classes, while minizing the false negatives. This is useful when missing positive cases (false negatives) will be harmful. The F1 score of both models is similar, though perceptron has a slightly better trade-off between accuracy and precision.

```python
def k_fold_cross_validation(X, y, model_type, n_splits=5):
    accuracies = []
    precisions = []
    recalls = []
    f1_scores = []
    auc_rocs = []

    X = X.to_numpy() if isinstance(X, pd.DataFrame) else X
    y = y.to_numpy() if isinstance(y, pd.Series) else y

    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

    for train_idx, val_idx in kf.split(X):
        X_train, X_val = X[train_idx], X[val_idx]
        y_train, y_val = y[train_idx], y[val_idx]

        scaler = StandardScaler()
        X_train = scaler.fit_transform(X_train)
        X_val = scaler.transform(X_val)

        if model_type == 'GaussianNB':
            model = GaussianNB()
        elif model_type == 'Perceptron':
            model = CustomPerceptron(learning_rate=0.01, n_iters=1000)

        model.fit(X_train, y_train)
        y_pred = model.predict(X_val)

        accuracy = accuracy_score(y_val, y_pred)
        precision = precision_score(y_val, y_pred)
        recall = recall_score(y_val, y_pred)
        f1 = f1_score(y_val, y_pred)



        accuracies.append(accuracy)
        precisions.append(precision)
        recalls.append(recall)
        f1_scores.append(f1)

    metrics = {
        'accuracy': np.mean(accuracies) * 100,
        'precision': np.mean(precisions) * 100,
        'recall': np.mean(recalls) * 100,
        'f1_score': np.mean(f1_scores) * 100,
    }



    return metrics
```

```python
X = df.drop('Outcome', axis = 1)
y = df['Outcome']
```

```python
n_splits = 5

metrics_nb = k_fold_cross_validation(X, y, 'GaussianNB', n_splits=n_splits)
print(f"Naive Bayes Cross-Validation Results (Averaged over {n_splits} folds):")
print(f"Accuracy: {metrics_nb['accuracy']:.3f}%")
print(f"Precision: {metrics_nb['precision']:.3f}%")
print(f"Recall: {metrics_nb['recall']:.3f}%")
print(f"F1 Score: {metrics_nb['f1_score']:.3f}%")
```

```python
n_splits = 5

metrics_per = k_fold_cross_validation(X, y, 'Perceptron', n_splits=n_splits)
print(f"Perceptron Cross-Validation Results (Averaged over {n_splits} folds):")
print(f"Accuracy: {metrics_per['accuracy']:.3f}%")
print(f"Precision: {metrics_per['precision']:.3f}%")
print(f"Recall: {metrics_per['recall']:.3f}%")
print(f"F1 Score: {metrics_per['f1_score']:.3f}%")
```

Both perceptron and naive bayes perform similar in terms of accuracy. Naive bayes has a slightly higher precision indicating that it is less likely to produce false positives compared to perceptron. Perceptron, however, has slightly better recall indicating that it is better at identifying positive class, with fewer false negatives. The F1 score of both models is quite similar, with that of Perceptron being only slightly higher.
